```python
# Dane produktów w sklepie
produkty = pd.DataFrame({
    'nazwa':      ['Laptop', 'Monitor', 'Klawiatura', 'Mysz'],
    'cena':       [3499, 1299, 249, 129],
    'stan_mag':   [12, 8, 45, 60],
    'kategoria':  ['Komputery', 'Komputery', 'Akcesoria', 'Akcesoria']
})
```

```python
# Filtrowanie – produkty droższe niż 500 zł
produkty[produkty['cena'] > 500]
```

Cofnijmy się jednak o krok. Co się dzieje w warunku porównania? Dostajemy Serię danych kolumny cena z wartościami True False. to tak jakby powiedzieć:
produkty[z kolumny cena weź te wieersze, gdzie warunek jest spełniony (True)]

```python
produkty['cena'] > 500
```

```python
# Filtrowanie z wieloma warunkami
produkty[(produkty['kategoria'] == 'Komputery') & (produkty['stan_mag'] > 8)]
```

wiemy, że True & True daje True i w tym filtrowaniu dzieje się dokładnie to samo! rodukty[ True & True = True (z dwóch warunków dadzą True i tylko te wiersze będą uwzglednione)]

```python
# Dodanie nowej kolumny – wartość magazynowa
produkty['wartosc_mag'] = produkty['cena'] * produkty['stan_mag']
produkty
```

na skutek dodania kolumny jest ona już uwzględniona w dataframe!

```python
# Kolumna z kategorią cenową
def kategoria_cenowa(cena):
    if cena >= 1000:
        return 'premium'
    elif cena >= 300:
        return 'srednia'
    else:
        return 'budzetowa'

produkty['segment'] = produkty['cena'].apply(kategoria_cenowa)
print(produkty)
```

```python
# Sortowanie po wartości magazynowej
#https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.sort_values.html
produkty.sort_values('wartosc_mag', ascending=False)
```

```python
produkty.sort_values('wartosc_mag')
```

```python
# Podstawowe statystyki
print("Średnia cena:   ", produkty['cena'].mean())
print("Odch. std. ceny:", produkty['cena'].std())
print()
print(produkty.describe())
```

```python
# Definicja prostej funkcji
def oblicz_vat(cena_netto, stawka=0.23):
    vat = cena_netto * stawka
    brutto = cena_netto + vat
    print(f"Cena netto: {cena_netto} zł | VAT: {vat:.2f} zł | Brutto: {brutto:.2f} zł")
    return brutto

oblicz_vat(500)
oblicz_vat(500, stawka=0.08)  # stawka obniżona
```

```python
### ZADANIE ###
# Stwórz funkcję, która przyjmie listę cen produktów i zwróci ich średnią
```

```python


def srednia_cen(ceny):
    return sum(ceny) / len(ceny)

ceny_produktow = [199, 349, 89, 499, 129]
print(srednia_cen(ceny_produktow))
```

```python
# Funkcja z wartością domyślną
def powitanie_klienta(imie='Kliencie', godzina=12):
    if godzina < 12:
        pora = 'Dzień dobry'
    elif godzina < 18:
        pora = 'Dzień dobry'
    else:
        pora = 'Dobry wieczór'
    return f"{pora}, {imie}!"

print(powitanie_klienta())
print(powitanie_klienta('Marta', godzina=19))
```

```python
# Funkcja z warunkami – ocena rentowności
def ocena_rentownosci(marza):
    if marza >= 0.30:
        return 'wysoka'
    elif marza >= 0.15:
        return 'średnia'
    else:
        return 'niska'

print(ocena_rentownosci(0.35))
print(ocena_rentownosci(0.20))
print(ocena_rentownosci(0.08))
```

```python
# Funkcja z dwoma zmiennymi
def zysk(przychod, koszt):
    return przychod - koszt

zysk(150000, 98000)
```

## 5. PĘTLE

```python
# Pętla for – iteracja po zakresie
for miesiac in range(1, 7):
    print(f"Miesiąc: {miesiac}")
```

```python
# Iteracja po liście elementów
dzialy = ['Sprzedaż', 'Marketing', 'Finanse', 'IT']
for dzial in dzialy:
    print(f"Dział: {dzial}")
```

```python
# Pętla while – symulacja spłaty pożyczki
saldo = 10000
rata = 2500
miesiac = 1

while saldo > 0:
    saldo -= rata
    if saldo < 0:
        saldo = 0
    print(f"Miesiąc {miesiac}: pozostałe saldo = {saldo} zł")
    miesiac += 1
```

```python
# break i continue
#https://www.w3schools.com/python/ref_func_range.asp
for i in range(1, 10):
    if i == 4:
        continue   # pomija i=4
    if i == 7:
        break      # kończy pętlę przy i=7
    print(f"i = {i}")
```

```python
### ZADANIE 3 ###
# Masz listę transakcji. Oblicz sumę tylko tych,
# których wartość przekracza 1000 zł.
```

```python


transakcje = [450, 1200, 890, 2300, 750, 1800, 300]
suma_duzych = 0

for t in transakcje:
    if t > 1000:
        suma_duzych += t

print(f"Suma dużych transakcji: {suma_duzych} zł")

# Elegantszy sposób – składnia list comprehension:
suma_duzych2 = sum(t for t in transakcje if t > 1000)
print(f"Wynik (list comprehension): {suma_duzych2} zł")
```
