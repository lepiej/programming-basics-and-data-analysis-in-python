```python
dane = ["123", "abc", None, "45"]

for element in dane:
    try:
        wynik = int(element) * 3
        print(wynik)
    except (TypeError, ValueError):
        print(f"Nie można przetworzyć: {element}")
```

```python
start = 45.20
end = 68.75

stopa_zwrotu = ((end - start) / start) * 100

print(f"Stopa zwrotu: {stopa_zwrotu:.2f}%")
```

```python
import pandas as pd

spolki = pd.DataFrame({
    'nazwa': ['Allegro', 'KGHM', 'Pepco', 'PKO BP'],
    'cena': [31.0, 155.0, 30.0, 85.0],
    'wolumen': [500000, 300000, 200000, 800000]
})

# filtrowanie
wynik = spolki[spolki['cena'] > 100]

print(wynik)
```
